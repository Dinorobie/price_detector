```python
# Shaping the dataframe to be able to merge
data04['precio_m2'] = data04['price'] / data04['building_size']
data04.rename(columns = {'building_size': 'superficie_total', 'category': 'categoria', 'bedrooms': 'dormitorios',
                         'bathrooms': 'baños', 'car_spaces': 'cocheras'}, inplace = True)
data04.drop(columns = ['land_size', 'rooms_total', 'furnished', 'land_size', 'id', 'subcategory', 'year_built', 'description',
                      'link', 'price'], inplace = True)

data04
```

```python
# Turning location into a string
location = data04['location'].tolist()
failed = 0
succes = 0
colonia = []
alcaldia = []
for x in range(0, len(location)):
    try:
        col, alc = location[x].split(', ')
        succes +=1
    except:
        col, alc = 'unknown', 'unknown'
        failed += 1
    colonia.append(col)
    alcaldia.append(alc)
print (f'Succesful: {succes} Failed: {failed}')
```

```python
# Formatting colonia
colonia = [unidecode.unidecode(x.upper()) for x in colonia]
len(colonia)
```

```python
# Formatting alcaldia
alcaldia = [unidecode.unidecode(x.upper()) for x in alcaldia]
len(alcaldia)
```

```python
# Appending to dataframe
data04['colonia'] = colonia
data04['alcaldia'] = alcaldia
data04.drop(columns = ['location'], inplace = True)
data04
```
