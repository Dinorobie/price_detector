```python
total_data={}
for i in range(0,page) :
    counter = counter + 1 
    counter_str = str(counter)
    url = 'https://www.lamudi.com.mx/distrito-federal/casa/for-sale/?currency=mxn&page=' + counter_str
    browser.visit(url)
    html = browser.html
    soup = bs(html, 'html.parser')
    
    features = soup.find_all('div', class_="ListingCell-AllInfo ListingUnit")
    description=[]
    data={}
    for i in features:
        extra_features=soup.find_all('div', class_="ListingCell-shortDescription")
        for i in extra_features:
            d=i.text
            description.append(d)
            
    for i in range(0,90):
        feature_data={
            "bathrooms":[],
            "bedrooms":[],
            "rooms_total":[],
            "building_size":[],
            "car_spaces":[],
            "category":[],
            "furnished":[],
            "location":[],
            "land_size":[],
            "price":[],
            "id":[],
            "subcategory":[],
            "year_built": [],
            "description":[]
        }
        try:
            feature_data["bathrooms"]=features[i]['data-bathrooms']
        except:
            print("variable not found")
        try:    
            feature_data["bedrooms"]=features[i]['data-bedrooms'],
        except:    
             print("variable not found")
        try:
            feature_data["rooms_total"]=features[i]['data-rooms_total'],
        except:
            print("variable not found")
        try:    
            feature_data["building_size"]=features[i]['data-building_size'],
        except:
            print("variable not found")
        try:
            feature_data["car_spaces"]=features[i]['data-car_spaces'],
        except:
            print("variable not found")
        try:
            feature_data["category"]=features[i]['data-category'],
        except:
            print("variable not found")
        try:
            feature_data["furnished"]=features[i]['data-furnished'],
        except:
            print("variable not found")
        try:
            feature_data["location"]=features[i]['data-geo-point'],
        except:
            print("variable not found")
        try:
            feature_data["land_size"]=features[i]['data-land_size'],
        except:
            print("variable not found")
        try:
            feature_data["price"]=features[i]['data-price'],
        except:
            print("variable not found")
        try:
            feature_data["id"]=features[i]['data-sku'],
        except:
            print("variable not found")
        try:
            feature_data["subcategory"]=features[i]['data-subcategories']
        except:
            print("variable not found")
        try:
            feature_data["year_built"]=features[i]['data-year_built']
        except:
            print("variable not found")
        try:
            feature_data["description"]=description[i]
        except:
            print("variable not found")
            
        
        data[i]=feature_data
    total_data[i]=data

total_data
```
